```python
input_file = open("romeoandjuliet.txt","r") 
```

```python
dict_words = {}
for line in input_file:
    for i in line.split():
        if i not in dict_words.keys():
            dict_words[i] = 1
        else:
            curr = dict_words[i]
            dict_words[i] = curr + 1
print(dict_words)
```

## Plot the bar graph for the frequency of all the words in the text:

```python
import matplotlib.pyplot as plt

plt.bar(range(len(dict_words)), list(dict_words.values()), align='center')
plt.xticks(range(len(dict_words)), list(dict_words.keys()), rotation=45, fontsize=8)
plt.xlabel('Words')
plt.ylabel('Frequency of Words')
plt.show()
```

## To see the top 10 most frequent words: We cut the bar graph to only look at the top 20 words:


```python
d_descending = sorted(dict_words, key=dict_words.get, reverse=True)
counter = 0
new_dict = {}
for r in d_descending:
    if counter == 20:
        break
    new_dict[r] = dict_words[r]
    counter += 1
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(new_dict)), list(new_dict.values()), align='center')
plt.xticks(range(len(new_dict)), list(new_dict.keys()), rotation=45, fontsize=8)
plt.xlabel('Words')
plt.ylabel('Frequency of Words')
plt.show()
```

Thus, we can see from the bar graph that the top 10 common words used in "romeoandjuliet.txt" are "the", "I", "and", "to", "a", "of", "my", "is", "in", "that". These words are either pronouns or common words like articles we need to help form a sentence.

## To see the least frequent 10 words, I cut the bar graph to look at only the least 20 frequent words:

```python
d_asc = sorted(dict_words, key=dict_words.get)
counter_asc = 0
new_dict_asc = {}
for i in d_asc:
    if counter_asc == 20:
        break
    new_dict_asc[i] = dict_words[i]
    counter_asc += 1
```

```python
import matplotlib.pyplot as plt

plt.bar(range(len(new_dict_asc)), list(new_dict_asc.values()), align='center')
plt.xticks(range(len(new_dict_asc)), list(new_dict_asc.keys()), rotation=45, fontsize=8)
plt.xlabel('Words')
plt.ylabel('Frequency of Words')
plt.show()
```

As we see from the bar graph, the least common 10 words all have frequency of appearing one time in the whole text file. They are: "Guards", "Hages", "Torchbearers", "houses:", "Gentlemen", "Sampson", "Abram,", "Laurence,"family.", "Man,". These words are all either actual names of people or names of objects.
